## 图像轮廓的检测
轮廓是一个整体，而边缘是零散的
* RETR_EXTERNAL：只检测最外边轮廓
* RETR_LIST：检索所有轮廓，将其保存至链表
* RETR_CCOMP：检索所有轮廓，组织为两层，顶层是各部分的外部边界，第二层是空洞的边界
* RETR_TREE：检索所有的轮廓，并重构嵌套轮廓的整个层次
轮廓逼近方法
* CHAIN_APPROX_NONE:以FREEMAN链码方式输出轮廓，所有其它方式输出多边形
* CHAIN_APPROX_SIMPLE:压缩水平的，垂直的和斜的部分，只保留终点部分

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

img=cv2.imread(r'C:\Users\MI\Desktop\My_picture\my_picture\my_picture..jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
cv_show('name',thresh)
ret
```

```python
contours,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)

draw_img=img.copy()
res=cv2.drawContours(draw_img,contours,-1,(0,0,255),2)
cv_show('name',res)
```

**轮廓特征**

```python
cnt=contours[0]
#面积
cv2.contourArea(cnt)
#周长True表示闭合的
cv2.arcLength(cnt,True)
```

```python
cnt=contours[0]
epsilon=0.1*cv2.arcLength(cnt,True)
approx=cv2.approxPolyDP(cnt,epsilon,True)

draw_img=img.copy()
res=cv2.drawContours(draw_img,[approx],-1,(0,255,0),2)
cv_show('name',res)
```

# 傅里叶变换
所有的周期图可以使用正弦波叠加
频域（横坐标师频率，纵坐标师振幅，描述了一个信号可以由哪些波组成）
高频：变化剧烈的灰度分量，例如边界
低频：变化缓慢的灰度分量
**滤波*
低通滤波器：只保留低频滤波器
高通滤波器：只保留高频滤波器

* 输入图像应转换为np.float32格式
* 得到结果频率为0的部分会在左上角，通常要转换到中心位置，可以通过shift变换来实现
* cv2.dift返回结果师双通道的（实部，虚部），通常要转换为图像模式才能展示

```python
#cv_show('name',img)
img_0=cv2.resize(img,(500,500))
cv_show('name',img_0)
```

```python
img_float32=np.float32(img)
dft=cv2.dft(img_float32,flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift=np.fft.fftshift(dft)

magnitude_spectrum=20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('Input Image'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum,cmap='gray')
plt.title('Mahnitude Spectrum'),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img=cv2.imread(r'C:\Users\MI\Desktop\My_picture\my_picture\face.jpg',0)
img=cv2.resize(img,(500,500))

img_float32=np.float32(img)

dft=cv2.dft(img_float32,flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift=np.fft.fftshift(dft)

rows,cols=img.shape
crow,ccol=int(rows/2),int(cols/2)  #中心位置

#低波通道
mask=np.zeros((rows,cols,2),np.uint8)
mask[crow-30:crow+30,ccol-30:ccol+30]=1

#IDFT
fshift=dft_shift*mask
f_ishift=np.fft.ifftshift(fshift)
img_back=cv2.idft(f_ishift)
img_back=cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('Input Image'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(img_back,cmap='gray')
plt.title('Result'),plt.xticks([]),plt.yticks([])

plt.show()
```
